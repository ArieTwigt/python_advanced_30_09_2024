```python
selected_columns = ['purpose', 'credit_amount', 'age_years', 'duration_months']

data_grouped = (
    data[selected_columns]
    .groupby('purpose')
    .agg({
        'credit_amount': 'sum',
        'age_years': 'mean',
        'duration_months': 'max'
    })
    .assign(age_years = lambda x: x['age_years'].round())
    .rename(columns={
        'credit_amount': 'Credit Amount',
        'age_years': 'mean_age_years',
        'duration_months': 'max_duration_months'
    })
    .sort_values("mean_age_years", ascending=False)
    .reset_index()
    

)

data_grouped
```

```python
data_grouped.to_csv("data/export/loan_purposes.csv", index=False)
```

```python
# export to excel
with pd.ExcelWriter('data/export/loans.xlsx') as writer:  
    # overview loans
    data.to_excel(writer,
                  sheet_name="Overview",)

    # loan purposes
    data_grouped.to_excel(writer, 
                          sheet_name="Loan Purposes")
```
